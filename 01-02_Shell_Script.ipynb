### Special Variable

```python
%%bash

# Script name
echo $0
# First input
echo $1
# number of inputs
echo $#
# All inputs
echo $*
set -- "arg  1" "arg  2" "arg  3"
for word in "$@"; do echo "$word"; done
# 0 --> no error, 1 --> error
echo $?
```
